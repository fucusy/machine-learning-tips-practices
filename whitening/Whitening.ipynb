```lua
require 'nn';
require 'gnuplot';
Plot = require 'itorch.Plot';
```

```lua
point_count = 100
random2DData = torch.randn(point_count, 1, 2)
random2DData[{{}, {}, {2}}] = torch.add(torch.mul(random2DData[{{}, {}, {1}}], 10), torch.randn(point_count, 1, 1))


selectX = torch.totable(random2DData[{{}, {}, 1}])
selectY = torch.totable(random2DData[{{}, {}, 2}])
```

```lua
plot = Plot():circle(selectX, selectY, 'red', 'hi'):draw()
plot:title('Scatter Plot Demo'):redraw()
plot:xaxis('x'):yaxis('y'):redraw()
plot:legend(true):redraw()
plot:save('before_whitening.html')
```

# start whiteninig preprocessing

```lua
std = torch.std(random2DData, 1)
mean = torch.mean(random2DData, 1)
print('previous std ')
print(std)
print('previous mean ')
print(mean)

mean = mean:expandAs(random2DData)
whitening2DData = torch.add(random2DData, -mean)
whitening2DData[{{}, {}, {1}}]:mul(1/std[1][1][1])
whitening2DData[{{}, {}, {2}}]:mul(1/std[1][1][2])


after_std = torch.std(whitening2DData, 1)
after_mean = torch.mean(whitening2DData, 1)
print('after std')
print(after_std)
print('after mean')
print(after_mean)
```

```lua
afterSelectX = torch.totable(whitening2DData[{{}, {}, {1}}])
afterSelectY = torch.totable(whitening2DData[{{}, {}, {2}}])
```

```lua
plot = Plot():circle(afterSelectX, afterSelectY, 'red', 'hi'):draw()
plot:title('Scatter Plot Demo'):redraw()
plot:xaxis('x'):yaxis('y'):redraw()
plot:legend(true):redraw()
plot:save('after_whitening.html')
```
